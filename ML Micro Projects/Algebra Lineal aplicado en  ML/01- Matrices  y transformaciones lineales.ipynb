# Descomposicin de matrices

```python
A = np.array([[-1,3], [2,-2]])
print(A)
```

```python
autovalores, autovectores = np.linalg.eig(A)
```

```python
print(autovalores)
print(autovectores)
```

```python
A_calc = autovectores.dot(np.diag(autovalores)).dot(np.linalg.inv(autovectores))
```

```python
print(A_calc)
```

```python
A = np.array([[3,2], [2,3]])
print(A)
```

```python
print(A == A.T)
```

```python
autovalores, autovectores = np.linalg.eig(A)
print(autovalores)
print(autovectores)
```

```python
A_calc = autovectores.dot(np.diag(autovalores)).dot(autovectores.T)
print(A_calc)
```

# Interpretacin de los valores singulares

```python
A = np.array([[3,7], [5,2]])
print(A)
```

```python
U, D, V = np.linalg.svd(A)

print(D[0])
print(D[1])
```

```python
u1 = [D[0]*U[0,0], D[0]*U[0,1]]
v1 = [D[1]*U[1,0], D[1]*U[1,1]]

print([A[0,0], A[1,0]])
print(u1)
print()
print([A[0,1], A[1,1]])
print(v1)
```

```python
graficarMatriz(A)
graficarVectores([u1,v1], cols =["red", "blue"])

plt.text(3,5, r"$u_1~$", size=18)
plt.text(7,2, r"$v_1$", size=18)

plt.text(-5,-4, r"$D(u_1)$", size=18)
plt.text(-4,1, r"$D(v_1)$", size=18)
```
